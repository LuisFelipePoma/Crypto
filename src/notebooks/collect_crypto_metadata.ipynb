# Using CoinMarketAPI for `eda_graph`


## Obtener las categorias

```python
# Obtenemos la lista de criptomonedas
categories_coinmarket = fetch_data(
    api=APIS.COINMARKET, uri="cryptocurrency/categories", parameters={}
)
# Obtenemos la data de las categorias
categories_coinmarket = categories_coinmarket["data"]
```

```python
# Filtrar los portafolios utilizando la expresión regular
regex_portfolios = create_regex(["portfolio"])
portfolios_coinmarket = list(
    filter(
        lambda categoria: regex_portfolios.search(categoria["title"].lower()),
        categories_coinmarket,
    )
)

# Filtrar las categorías utilizando la expresión regular
CATEGORIES_WORDS: list[str] = [
    "meme",
    "memes",
    "game",
    "gaming",
    "ai",
    "real world assets",
]

# Filtrar las categorías utilizando la expresión regular
regex_categories = create_regex(words=CATEGORIES_WORDS)
categories_coinmarket = list(
    filter(
        lambda categoria: regex_categories.search(categoria["title"].lower()),
        categories_coinmarket,
    )
)
```

```python
# Guardar las categorias filtradas
save_json(categories_coinmarket, "../notebooks/eda_graph/metadata/categories.json")

# Guardar los portafolios filtrados
save_json(portfolios_coinmarket, "../notebooks/eda_graph/metadata/portfolios.json")
```

## Obtener los exchanges

```python
exchanges_gecko = []
page = 1
while True:
    exchange = fetch_data(
        api=APIS.COINGECKO, uri=f"exchanges?page={page}", parameters={}
    )
    if not exchange: # Si no hay más exchanges
        break
    exchanges_gecko.append(exchange)
    page += 1
    time.sleep(1)
```

```python
# lista de 2 dimensiones a 1 dimension
exchanges_gecko = [item for sublist in exchanges_gecko for item in sublist]
```

```python
len(exchanges_gecko)
```

```python
# Guardar las exchanges metadata
save_json(exchanges_gecko, "../notebooks/eda_dataset/metadata/exchanges.json")
```
