Visualizacion de las coins mas caras segun su precio actual

```python
# visualizacion con plt de las current_price de las 10 coins mas caras
df_pivot = df.sort_values('current_price', ascending=False).head(20)
plt.figure(figsize=(10, 5))
sns.barplot(x='current_price', y='name', data=df_pivot, palette='viridis',hue='name')
plt.title('Top 10 coins segun su precio actual')
# cambiar la notacion cientifica de los ejes a numeros
plt.ticklabel_format(style='plain', axis='x')
plt.xlabel('Precio actual')
plt.ylabel('Nombre')
plt.show()
```

Podemos observar que actualmente la coin con mayor precio actual de mercado es `Zorksees` con un valor mayor a 1,4 millones de dolares, lo que indica una alta capitalizaciÃ³n y potencial interÃ©s por parte de los inversores. Por otro lado, los demas no sobrepasan el valor de los 400 mil dolares, sugiriendo una distribuciÃ³n variada en los precios y capitalizaciones dentro del mercado de criptomonedas.

Visualizacion de las coins segun su capitalizacion de mercado

```python
# visualizacion con plt de las coins mas populares
df_pivot = df.sort_values('market_cap', ascending=False).head(20)
plt.figure(figsize=(10, 5))
sns.barplot(x='market_cap', y='name', data=df_pivot, palette='viridis',hue='name')
plt.title('Top 10 coins segun su capitalizacion de mercado')
# cambiar la notacion cientifica de los ejes a numeros
plt.ticklabel_format(style='scientific', axis='x', useMathText= True, scilimits=(6,6))

plt.xlabel('Capitalizacion de mercado')
plt.ylabel('Nombre')
plt.show()
```

Podemos observar que actualmente `Bitcoin` es la moneda con mayor capitalizacion de mercado (1 billon de dolares) por una amplia diferencia con respecto del segundo que es `Ethereum`, mientras que las demas coins no sobrepasan 200 mil millones de dolares. Aqui se destaca el claro liderazgo de Bitcoin en el mercado de criptomonedas, subrayando su posiciÃ³n preeminente en comparaciÃ³n con otras monedas digitales.

Visualizacion de las monedas con el precio historico mas alto

```python
# visualizacion con plt de las coins con el precio historico mas alto
df_pivot = df.sort_values('ath', ascending=False).head(n=10)
plt.figure(figsize=(10, 5))
sns.barplot(x='ath', y='name', data=df_pivot, palette='viridis',hue='name')
plt.title('Top 10 coins segun su precio historico mas alto')
# cambiar la notacion cientifica de los ejes a numeros
plt.ticklabel_format(style='scientific', axis='x', useMathText= True, scilimits=(5,5))
plt.xlabel('Precio historico mas alto')
plt.ylabel('Nombre')
plt.show()
```

El grafico de barras nos da una vision de las monedas que obtuvieron el mayor precio historico, destacandose `Calaxy` con casi 60 millones de dolares en su maximo historico, siguiendole `Peepo (SOL)` con 30 millones de dolares y los demas no sobrepasando los 100 millones de dolares. Se resalta las monedas que han alcanzado valores significativos en el pasado, destacando la notable diferencia entre las principales criptomonedas y el resto en tÃ©rminos de precios histÃ³ricos mÃ¡ximos.

Visualizacion de las fechas de los maximos historicos

```python
df_pivot = df.copy()
df_pivot['ath_date'] = pd.to_datetime(df_pivot['ath_date'])

plt.figure(figsize=(10, 5))
sns.kdeplot(df_pivot['ath_date'], color='purple', fill=True) # type: ignore
plt.title('Densidad de las fechas de los maximos historicos')
plt.xlabel('Fecha')
plt.ylabel('Densidad')
plt.show()
```

La grafica nos muestra la densidad de las fechas en las que nuestras monedas en el dataset obtuvieron su maximo historico. Como se puede observar, los aÃ±os que hubieron picos de crecimiento fueron durante los aÃ±os `2018`, `2022` y `2024`, siendo el ultimo el de mayor densidad. Tambien nos es Ãºtil para identificar tendencias y patrones de crecimiento en el mercado a lo largo del tiempo.

Visualizacion de las monedas que presentaron mayor cambio en su precio las 24 horas ultimas

```python
# Visualizacion de las monedas que presentaron mayor cambio en su precio las 24 horas ultimas
df_pivot = df.sort_values('price_change_percentage_24h', ascending=False).head(25)
plt.figure(figsize=(10, 5))
sns.pointplot(x='price_change_percentage_24h', y='name', data=df_pivot, palette='viridis', hue='name')
plt.title('Coins con mayor cambio de precio en las ultimas 24 horas')
plt.xlabel('Cambio de precio en las ultimas 24 horas')
plt.ylabel('Nombre')
plt.show()
```

El grafico nos muestra las monedas que mas cambio han sufrido en las ultimas 24 horas, teniendo a `McPepe's` que sufrio el mayor cambio de casi mas del 2500%, siguiendole `Monero-Classic` con casi 1000%, `Relay Chain` y `PEPE UWU` con casi 500%, mientras las demas monedas mostraron un cambio no mayor del 250% aproximadamente. Estos datos reflejan la volatilidad y los cambios extremos en los precios de algunas criptomonedas en un perÃ­odo de 24 horas, destacando las monedas mÃ¡s afectadas en tÃ©rminos de fluctuaciones de precios durante este perÃ­odo especÃ­fico.
