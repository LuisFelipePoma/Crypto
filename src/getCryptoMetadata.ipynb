# Using CoinMarketAPI 

## Usamos la documentacion para obtener las categorias de Coin Market API

```python
from requests import Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json

url = "https://pro-api.coinmarketcap.com/v1/"

headers = {
    "Accepts": "application/json",
    "X-CMC_PRO_API_KEY": os.environ.get("API_KEY_COINMARKET"),
}


def fetch_data(uri: str, parameters: dict) -> dict:
    session = Session()
    session.headers.update(headers)
    try:
        response = session.get(url + uri, params=parameters)
        return json.loads(response.text)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
        return {"error": e}
```

```python
# Obtenemos la lista de criptomonedas
categories = fetch_data("cryptocurrency/categories", {})
# Obtenemos la data de las categorias
categories_data = categories["data"]
```

```python
import re

# Las categorias a buscar
categories_names = ["memes", "game", "gaming", "ai", "real world assets"]

# Crear una expresión regular que coincida con cualquier palabra clave
regex_pattern = r"\b(?:{})\b".format("|".join(categories_names))
regex = re.compile(regex_pattern, re.IGNORECASE)

# Filtrar las categorías utilizando la expresión regular
categories_filter = list(
    filter(lambda categoria: regex.search(categoria["title"].lower()), categories_data)
)
```

```python
# Crear un directorio para guardar los metadatos
os.makedirs("metadata", exist_ok=True)

# Guardamos las categorias filtradas
with open("metadata/categories_coinmarket.json", "w") as file:
    json.dump(categories_filter, file)
```
