# Using CoinMarketAPI


## Obtener las categorias

```python
# Obtenemos la lista de criptomonedas
categories_coinmarket = fetch_data(
    api=APIS.COINMARKET, uri="cryptocurrency/categories", parameters={}
)
# Obtenemos la data de las categorias
categories_coinmarket = categories_coinmarket["data"]
```

```python
# Filtrar los portafolios utilizando la expresión regular
regex_portfolios = create_regex(["portfolio"])
portfolios_coinmarket = list(
    filter(
        lambda categoria: regex_portfolios.search(categoria["title"].lower()),
        categories_coinmarket,
    )
)

# Filtrar las categorías utilizando la expresión regular
regex_categories = create_regex(["memes", "game", "gaming", "ai", "real world assets"])
categories_coinmarket = list(
    filter(
        lambda categoria: regex_categories.search(categoria["title"].lower()),
        categories_coinmarket,
    )
)
```

```python
# Guardar las categorias filtradas
save_json(categories_coinmarket, "metadata/coinmarket/categories_coinmarket.json")

# Guardar los portafolios filtrados
save_json(portfolios_coinmarket, "metadata/coinmarket/portfolios_coinmarket.json")
```

## Obtener las exchanges

```python
# Obtenemos la lista de criptomonedas
exchanges_coinmarket = fetch_data(
    api=APIS.COINMARKET, uri="exchange/map", parameters={}
)
# Obtenemos la data de las categorias
exchanges_coinmarket = exchanges_coinmarket["data"]
```
